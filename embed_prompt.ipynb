```python
import openai
import json
import os

def generate_answer(message, title, source, start_id):
    model="gpt-4o-mini"
    messages = [
        {
            "role": "system",
            "content": (
                "split the given text to be used to create json elements for embedding using the text-embedding-3-small model for a RAG chatbot."
                "use this json format: id, source, title, content."
                f"start from the given {start_id} and increment by 1 for each chunk."
                "ensure that content  length is between 700 and 1000 character."
                "IMPORTANT: while chunking pay attention to context preserving and character limit."
                "dont skip words, just use them as they are in the original text, dont change any letter."
                "write each json element in a new line as a single json object"
            ),
        },
        {
            "role": "user",
            "content": f"text:{message}. tilte:{title}. source:{source}",
        },
    ]
    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message.content.strip()
```

```python
chunks = []

if os.path.exists("chunks.json"):
    with open("chunks.json", "r", encoding="utf-8") as f:
        try:
            chunks = json.load(f)
        except json.JSONDecodeError:
            print("⚠️ Existing chunks.json is invalid JSON, starting fresh.")

start_id = len(chunks) + 1 

raw_output = generate_answer(message4, title4, source4, start_id=start_id)
print("answer generated")
print(raw_output)

lines = raw_output.splitlines()


for line in lines:
    try:
        chunks.append(json.loads(line))
    except json.JSONDecodeError:
        print("⚠️ Skipped invalid JSON line:", line)

with open("chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, ensure_ascii=False, indent=4)

print(f"✅ Saved {len(chunks)} total chunks to chunks.json")
```
